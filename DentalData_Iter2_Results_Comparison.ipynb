## Functions to plot the results

```python
import matplotlib.pyplot as plt

def plot_one(data: list[float], title: str):
    x = range(len(data))
    plt.plot(x, data)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(title)
    plt.show()

def plot_all(data_list: list[list[float]], labels: list[str], title: str):
    x = range(len(data_list[0]))
    for label, data in zip(labels, data_list):
        plt.plot(x, data, label=label)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(title)
    plt.legend()
    plt.show()
```
